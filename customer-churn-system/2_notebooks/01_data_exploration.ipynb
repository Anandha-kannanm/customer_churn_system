```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

ROOT = Path('..').resolve()
df = pd.read_csv(ROOT / '1_data' / 'raw' / 'telco_churn.csv')
df.info()
df.describe()
```

```python
sns.countplot(data=df, x='Churn')
plt.title('Churn Distribution')
plt.show()
```
